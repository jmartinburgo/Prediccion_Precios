```python
#Construimos el modelo
x=df_dummies.drop('price',axis=1)
x
```

```python
y=df_dummies['price']
y
```

```python
  from sklearn.model_selection import train_test_split

  x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=7)
```

```python
#Estandarizamos las variables independientes
from sklearn.preprocessing import StandardScaler

StandardScaler()
#Entremos el standar
scaler=StandardScaler().fit(x_train)

scaler
```

```python
#Transformo las varibles independientes de entrenamiento
x_train_scaler = scaler.transform(x_train)
x_train_scaler
```

```python
x_test_scaler = scaler.transform(x_test)
x_test_scaler
```

```python
#Modelo k Vecinos mas cercanos

from sklearn.neighbors import KNeighborsRegressor

knn=KNeighborsRegressor(n_neighbors=10)

knn.fit(x_train_scaler,y_train)
```

```python
#r2 del modelo

r2_knn= knn.score (x_test_scaler,y_test)
r2_knn
```

```python
#Modelo Arbol de decision
from sklearn.tree import DecisionTreeRegressor

tree= DecisionTreeRegressor(random_state=0)

tree.fit(x_train_scaler,y_train)
```

```python
#r2 del modelo
r2_tree=tree.score(x_test_scaler,y_test)

r2_tree
```

```python
#Modelo Random Forest

from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(random_state=5)

rf.fit(x_train_scaler,y_train)
```

```python
r2_rf=rf.score(x_test_scaler,y_test)
r2_rf
```

```python
#Modelo Gradient Boosting

from sklearn.ensemble import GradientBoostingRegressor

gb=GradientBoostingRegressor(random_state=10)

gb.fit(x_train_scaler,y_train)
```

```python
r2_gb=gb.score(x_test_scaler,y_test)
r2_gb
```

```python
#Modelo Gradient Boosting

from xgboost import XGBRegressor

xgb=XGBRegressor(random_state=48)

xgb.fit(x_train_scaler,y_train)
```

```python
r2_xgb=xgb.score(x_test_scaler,y_test)

r2_xgb
```

```python
models=pd.DataFrame({
    'Model':['KNN','TREE','RANDOM FOREST','GRADIENT BOOSTING','XGBOOSTING'],
    'Score':[r2_knn,r2_tree,r2_rf,r2_gb,r2_xgb]
}
)

models.sort_values(by='Score',ascending=False)
```

```python
#Prediccion con XGB

y_pred = xgb.predict(x_test_scaler)
y_pred
```

```python
#Evaluacion del modelo

from sklearn.metrics import mean_squared_error

from sklearn import metrics

```

```python
#r2
r2=metrics.r2_score(y_test,y_pred)
r2
```

```python
#r2 ajustado
r2_ajust=1-(1-r2)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
r2_ajust
```

```python
#MAE

MAE= metrics.mean_absolute_error(y_test,y_pred)
MAE
```

```python
#MAE

MSE= metrics.mean_squared_error(y_test,y_pred)
MSE
```

```python
#RMSE

RMSE=np.sqrt(MSE)
RMSE
```
